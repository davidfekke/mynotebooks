## Section 3: Return Statements

Here's the key difference:
- **`print()`**: Shows output to the screen (for humans to see)
- **`return`**: Sends a value back so your code can use it later

**Why This Matters in AI/ML:**
When building models, you need to RETURN values (like accuracy scores, predictions, processed data) so other parts of your code can use them. Printing is just for debugging and visualization.

````python
# TODO: Define add_and_print(a, b) - this one prints the sum
def add_and_print(a, b):
    # TODO: Calculate the sum and print it
    print(f"{a + b}")

# TODO: Define add_and_return(a, b) - this one returns the sum
def add_and_return(a, b):
    # TODO: Calculate the sum and return it
    return a + b

# Test both functions
print("Testing print version:")
result1 = add_and_print(5, 3)
print(f"Stored result: {result1}")  # What value is stored?

print("\nTesting return version:")
result2 = add_and_return(5, 3)
print(f"Stored result: {result2}")  # What value is stored?
````

## Section 4: Variable Scope

**Scope = Where variables can be accessed in your code**

- **Local scope**: Variables created inside a function - only accessible within that function
- **Global scope**: Variables created outside functions - accessible everywhere in your code

**Analogy:** Think of it like rooms in a house:
- 🛏️ Local variables = Bedroom items (only accessible in that room)
- 🍳 Global variables = Kitchen (everyone in the house can access it)

### Exercise 5.2: Powers and Remainders

**Task:** Create a function called `process_data_batch(total_items, batch_size)` that:
1. Calculates how many full batches you can create
2. Calculates how many items are left over
3. Returns both values

**Real-World Context:**
When training machine learning models, you often process data in batches. If you have 1000 images and a batch size of 32, you get 31 full batches with 8 images left over.

**Hint:** Use `//` (integer division) and `%` (modulo)

````python
# TODO: Define process_data_batch(total_items, batch_size)
def process_data_batch(total_items, batch_size):
    # TODO: Calculate full_batches using integer division (//)
    
    full_batches = total_items // batch_size  # Replace with your calculation
    # TODO: Calculate leftover_items using modulo (%)
    leftover_items = total_items - (full_batches * batch_size)  # Replace with your calculation
    
    return full_batches, leftover_items

# TODO: Test with 1000 items and batch size of 32
batches, leftover = process_data_batch(1000, 32)
print(f"Full batches: {batches}")
print(f"Leftover items: {leftover}")
````

````python
# TODO: Define complex_calculation(a, b, c)
def complex_calculation(a, b, c):
    # TODO: Implement the formula: (a + b) * c ** 2 / 10
    result = (a + b) * c ** 2 /10  # Replace with your formula
    return result

# TODO: Test with different values
answer = complex_calculation(3, 7, 2)
print(f"Result: {answer}")
# Expected: (3 + 7) * 2^2 / 10 = 10 * 4 / 10 = 4.0
````

### Challenge: Data Analysis Helper Function

**Task:** Create a function called `analyze_dataset(dataset_name, total_records, processed_records)` that:

1. Calculates the **percentage of records processed**
2. Calculates **how many records remain**
3. Returns a **formatted string** with all this information

**Example Output:**
```python
analyze_dataset("Customer Data", 1000, 750)
# Returns:
# "Dataset: Customer Data
#  Progress: 75.0% complete
#  Remaining: 250 records"
```

**Real-World Context:**
From my consulting work, I often run the same analysis on different datasets - customer data, sales data, inventory data, patient metrics. This kind of function lets me track progress across all of them with reusable code!

````python
# TODO: Define analyze_dataset(dataset_name, total_records, processed_records)
def analyze_dataset(dataset_name, total_records, processed_records):
    # TODO: Calculate percentage complete
    percentage = processed_records / total_records * 100  # Replace with your calculation
    
    # TODO: Calculate remaining records
    remaining = total_records - processed_records  # Replace with your calculation
    
    # TODO: Create a formatted string with all the info
    # Hint: Use triple quotes """ for multi-line strings
    report = f"""Dataset: {dataset_name}
Progress: {percentage}% complete
Remaining: {remaining} records"""
    
    return report

# TODO: Test with different datasets
print(analyze_dataset("Customer Data", 1000, 750))
print()
print(analyze_dataset("Sales Data", 5000, 4200))
print()
print(analyze_dataset("Patient Metrics", 850, 425))
````

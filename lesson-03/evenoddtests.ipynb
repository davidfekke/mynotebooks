```python
def evenoddlogic(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
```

```python
print(evenoddlogic(4))  # Should print "Even"
print(evenoddlogic(7))  # Should print "Odd"
```

```python
def get_letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"
    
letter_grade = get_letter_grade(85)
print(letter_grade)  # Should print "B"
```

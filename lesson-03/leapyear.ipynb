```python
def is_leap(year):
    """Returns True if the given year is a leap year, False otherwise."""
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    return False

print(is_leap(2020))  # True
print(is_leap(1900))  # False
print(is_leap(2100))  # False
```

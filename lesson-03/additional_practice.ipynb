### Challenge: Create Basic Functions
Practice the function syntax we learned today.

```python
# TODO: Create three functions for the school system

# Function 1: Convert a numeric score to a letter grade
# Requirements:
#   - Takes a score (0-100)
#   - Returns: A (90-100), B (80-89), C (70-79), D (60-69), F (<60)
#   - Name it appropriately

# Your code here...
def get_letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

print(f"91 would be {get_letter_grade(91)}")
print(f"85 would be {get_letter_grade(85)}")
print(f"76 would be {get_letter_grade(76)}")

# Function 2: Calculate the average of a list of grades
# Requirements:
#   - Takes a list of numeric grades
#   - Returns the average as a float
#   - Handle empty lists gracefully

# Your code here...
def avg_grade(values):
    return sum(values) / len(values)

print(f"Average grade for [75, 95, 85] would be {avg_grade([75, 95, 85])}")


# Function 3: Generate a personalized greeting
# Requirements:
#   - Takes a student's name
#   - Returns a welcome message
#   - Bonus: Add time of day (morning/afternoon/evening)

# Your code here...
def greeting(name, time_of_day = "afternoon"):
    print(f"Good {time_of_day} {name}!")

greeting("Sally", "morning")

# Test your functions with these (after you create them):
print(get_letter_grade(85))
print(avg_grade([90, 85, 78, 92]))
print(greeting("Alice"))
```

```python
# Final Challenge: Build a Complete Student Management System

# Requirements:
# 1. Create a system that can:
#    - Store multiple students with their info (name, grade, courses)
#    - Add new students
#    - Update student grades
#    - Search for students by name
#    - List all students in a specific course
#    - Calculate class averages
#    - Generate report cards

# 2. Use appropriate data structures:
#    - Consider what's best for each type of data
#    - Think about lookup speed and uniqueness

# 3. Create at least 3 functions to handle common operations

# 4. Include error handling for:
#    - Invalid grades (must be 0-100)
#    - Student not found scenarios
#    - Empty course lists

# 5. Bonus features:
#    - Track attendance
#    - Calculate GPA with weighted courses
#    - Generate honor roll list

# Start here - plan your approach first!
# What data structures will you use? Why?
# What functions will make your code cleaner?

# Your code here...
students = {}

def add_student(name, grade, courses = []):
    students[name] = { "student_name": name, "grade": grade, "courses": courses }
def set_grade(name, grade):
    students[name]["grade"] = grade

def get_student(name):
    return students[name]

def get_students_by_course(course):
    student_list = students.values()
    return list(filter(lambda stu: course in stu["courses"], student_list))

def calculate_class_avg():
    student_list = students.values()
    scores = list(map(lambda student: student["grade"], student_list))
    return sum(scores) / len(scores)

def report_cards():
    for student_id in students.keys():
        student_dict = students[student_id]
        print(f"student {student_id} with score of {student_dict["grade"]}")


add_student("Dave", 89, ["History", "Computer Science", "Mathematics"])
add_student("Tom", 72, ["History", "Computer Science", "Biology"])
add_student("Carol", 95, ["English", "Physics", "Mathematics"])

report_cards()

# temp = get_student("Tom")
set_grade("Dave", 98)
temp = get_students_by_course("Computer Science")
print(temp)

print(calculate_class_avg())
```

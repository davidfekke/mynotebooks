### Challenge 4: Processing All Ingredients
Use a for loop to process your ingredient list.

**ROTATE DRIVER:** Switch who's typing!

```python
ingredients = ["tomatoes", "pasta", "cheese", "basil", "olive oil"]

# TODO: Print each ingredient with a number (1. tomatoes, 2. pasta, etc.)
# Hint: Look up enumerate() or use a counter variable
print("Shopping List:")
# Your code here...
for index, ingredient in enumerate(ingredients, start=1):
    print(f"{index}. {ingredient}")

# TODO: Count how many ingredients have more than 5 letters
# Your code here...
ingredient_counter_plus5 = 0
for ingredient in ingredients:
    if len(ingredient) > 5:
        ingredient_counter_plus5 += 1

# Is there a way I can get a count of ingredients with more than 5 letters without a loop?
ingredient_counter_plus5 = sum(1 for ingredient in ingredients if len(ingredient) > 5)

print(f"Number of ingredients with more than 5 letters: {ingredient_counter_plus5}")

# TODO: Create a new list with all ingredients in UPPERCASE
# Your code here...
upper_ingredients = []
for ingredient in ingredients:
    upper_ingredients.append(ingredient.upper())
print(f"Uppercase ingredients: {upper_ingredients}")

# BONUS: Create a list of just the first 3 letters of each ingredient
short_list = []
for ingredient in ingredients:
    short_list.append(ingredient[:3])
print(f"Shortened ingredients (first 3 letters): {short_list}")

```

```python
# TODO: Create a dictionary with at least 5 recipes and their ratings (1-5 stars)
recipe_ratings = {
    "Pasta Carbonara": 4.5,
    "Chicken Tikka Masala": 4.8,
    "Cesar Salad": 3.9
}

# TODO: Calculate and print the average rating
# Hint: You'll need to loop through the dictionary and keep track of total and count
# Your code here...
average_rating = sum(recipe_ratings.values()) / len(recipe_ratings)

# TODO: Find and print the highest rated recipe
# Your code here...
highest_rated_recipe = max(recipe_ratings, key=recipe_ratings.get)
highest_rating = recipe_ratings[highest_rated_recipe]
print(f"Highest rated recipe: {highest_rated_recipe} with {highest_rating} stars")

# TODO: Print all recipes with 4+ stars as "Recommended"
print("\n⭐ Recommended Recipes:")
# Your code here...
for recipe, rating in recipe_ratings.items():
    if rating >= 4:
        print(f"- {recipe} ({rating} stars)")
```

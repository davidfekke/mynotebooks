## Exercise 1: How Similar Are These Words?

Cosine similarity measures how close two word vectors are in meaning. The score ranges from **-1** (opposite) to **1** (identical).

**Your task:** Before running the code, **predict** which pair will score highest and which will score lowest. Then compute the actual scores and see if your intuition was right!

**Function:** `glove.similarity("word1", "word2")` returns a float between -1 and 1

```python
# These word pairs range from very similar to totally unrelated.
# STEP 1: Before running, discuss with your group —
#         which pair do you think will score highest? lowest?

word_pairs = [
    ("king", "queen"),
    ("happy", "joyful"),
    ("car", "automobile"),
    ("dog", "cat"),
    ("computer", "banana"),
    ("run", "sprint"),
]

# STEP 2: Compute the similarity for each pair and store in a list
# Each entry should be a tuple: (word1, word2, score)
similarities = []

for word1, word2 in word_pairs:
    score = glove.similarity(word1, word2)  # YOUR CODE HERE — replace None with the similarity calculation
    similarities.append((word1, word2, score))

# Print results
print("Word Pair Similarities:\n")
for w1, w2, score in similarities:
    print(f"  {w1:12s} ↔ {w2:12s}  →  {score:.4f}")
```

```python
# ✅ Checkpoint — run this cell to verify your work

from numpy import float32


assert len(similarities) == 6, f"Expected 6 similarity scores, got {len(similarities)}"
assert all(s[2] is not None for s in similarities), "Some scores are still None — did you replace the placeholder?"
assert all(isinstance(s[2], float32) for s in similarities), "Scores should be floats — make sure you're using glove.similarity()"

scores_only = [s[2] for s in similarities]
assert max(scores_only) > 0.5, "Your highest score should be above 0.5"
assert min(scores_only) < 0.3, "Your lowest score should be below 0.3"

print("✅ All checks passed!")
print(f"\n   Highest: {max(scores_only):.4f}")
print(f"   Lowest:  {min(scores_only):.4f}")
print(f"\n🤔 Did 'dog' ↔ 'cat' score higher or lower than you expected?")
```

## Exercise 2: Word Math — Analogies

The coolest thing about word embeddings: you can do **math with words**!

The classic example: **king - man + woman = queen**

This works because the vector from "man" to "king" captures the concept of "royalty," and adding that same direction to "woman" lands near "queen."

**How to read the code:**

> "A is to B as C is to ___?"
>
> `glove.most_similar(positive=["B", "C"], negative=["A"])`
>
> This computes: **B - A + C = ?**

**Your task:** Run the guided example, then build 2 analogies of your own!

```python
# GUIDED EXAMPLE: "man is to king as woman is to ___?"
# king - man + woman = ???

result = glove.most_similar(positive=["king", "woman"], negative=["man"], topn=3)

print("man → king  as  woman → ???\n")
for word, score in result:
    print(f"   {word:15s}  (score: {score:.4f})")
```

```python
# YOUR TURN — Analogy 1
# Think of an analogy: "A is to B as C is to ___?"
# Examples to try:
#   "paris is to france as tokyo is to ___?"  → positive=["france", "tokyo"], negative=["paris"]
#   "doctor is to hospital as teacher is to ___?" → positive=["hospital", "teacher"], negative=["doctor"]

analogy_1 = glove.most_similar(
    positive=["france", "tokyo"],  # ✏️ FILL IN: B and C
    negative=["paris"],          # ✏️ FILL IN: A
    topn=3
)

# What did you expect the answer to be?
expected_1 = "japan"  # ✏️ FILL IN your prediction

print("Analogy 1 results:\n")
for word, score in analogy_1:
    print(f"   {word:15s}  (score: {score:.4f})")
print(f"\n   You expected: '{expected_1}'")
```

```python
# YOUR TURN — Analogy 2
# Try something different! Sports, food, animals, jobs — anything goes.

analogy_2 = glove.most_similar(
    positive=["football", "pitcher"],  # ✏️ FILL IN: B and C
    negative=["quarterback"],          # ✏️ FILL IN: A
    topn=3
)

expected_2 = "baseball"  # ✏️ FILL IN your prediction

print("Analogy 2 results:\n")
for word, score in analogy_2:
    print(f"   {word:15s}  (score: {score:.4f})")
print(f"\n   You expected: '{expected_2}'")
```

```python
# ✅ Checkpoint — run this cell to verify your work

assert analogy_1 is not None, "analogy_1 hasn't been computed — fill in the blanks above!"
assert analogy_2 is not None, "analogy_2 hasn't been computed — fill in the blanks above!"
assert len(analogy_1) > 0, "analogy_1 returned no results — check your words are in the vocabulary"
assert len(analogy_2) > 0, "analogy_2 returned no results — check your words are in the vocabulary"

print("✅ Both analogies computed!")
print(f"   Analogy 1 top answer: '{analogy_1[0][0]}'  (you expected: '{expected_1}')")
print(f"   Analogy 2 top answer: '{analogy_2[0][0]}'  (you expected: '{expected_2}')")
print(f"\n🤔 Did GloVe match your predictions? Why or why not?")
```

## Exercise 3: Odd One Out

GloVe can figure out which word doesn't belong in a group! It compares all the word vectors and finds the one that's farthest from the others.

**Function:** `glove.doesnt_match(["word1", "word2", "word3", "word4"])` returns the outlier as a string

**Your task:**
1. Run the example to see how it works
2. Create 2 puzzles of your own — one easy, one tricky
3. The tricky one should try to **fool** GloVe (pick an outsider that's subtly related)

```python
# GUIDED EXAMPLE: which word doesn't belong?

example_words = ["breakfast", "lunch", "dinner", "soccer"]
odd_word = glove.doesnt_match(example_words)

print(f"Words:   {example_words}")
print(f"GloVe says '{odd_word}' doesn't belong!")
```

```python
# PUZZLE 1 (Easy): Pick 3 related words + 1 obvious outsider

puzzle_1 = ["pc", "laptop", "mac", "bicycle"]  # ✏️ CHANGE THESE

odd_1 = glove.doesnt_match(puzzle_1)
print(f"Puzzle 1:  {puzzle_1}")
print(f"GloVe says '{odd_1}' doesn't belong!")
```

```python
# PUZZLE 2 (Tricky): Can you FOOL GloVe?
# Pick an outsider that's subtly related — like "hotel" among hospitals/clinics
# If GloVe picks the wrong word, you win!

puzzle_2 = ["colonel", "corporal", "sergeant", "teacher"]  # ✏️ CHANGE THESE

odd_2 = glove.doesnt_match(puzzle_2)
print(f"Puzzle 2:  {puzzle_2}")
print(f"GloVe says '{odd_2}' doesn't belong!")
print(f"\n👀 Did GloVe get it right, or did you fool it?")
```

```python
# ✅ Checkpoint — run this cell to verify your work

assert isinstance(odd_1, str), "Puzzle 1 didn't return a result — fill in your words!"
assert isinstance(odd_2, str), "Puzzle 2 didn't return a result — fill in your words!"
assert odd_1 in puzzle_1, "Something went wrong with puzzle 1"
assert odd_2 in puzzle_2, "Something went wrong with puzzle 2"

print("✅ Both puzzles created!")
print(f"   Puzzle 1: '{odd_1}' is the odd one out")
print(f"   Puzzle 2: '{odd_2}' is the odd one out")
print(f"\n🎯 Pick your best puzzle to share with the class!")
```

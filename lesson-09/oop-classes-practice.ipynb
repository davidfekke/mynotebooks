### Problem 1.3: Create a Rectangle Class with Methods

Create a `Rectangle` class with methods to calculate area and perimeter.

**Formulas:**
- Area = width * height
- Perimeter = 2 * (width + height)

```python
class Rectangle:
    """
    A class representing a rectangle.
    
    Attributes:
        width (float): The rectangle's width
        height (float): The rectangle's height
    
    Methods:
        area(): Returns width * height
        perimeter(): Returns 2 * (width + height)
    """
    
    def __init__(self, width, height):
        # Your code here
        self.width = width
        self.height = height
    
    def area(self):
        # Your code here
        return self.width * self.height
    
    def perimeter(self):
        # Your code here
        return (self.width + self.height) * 2
```

```python
# Test cases - DO NOT MODIFY
rect1 = Rectangle(10, 5)
assert rect1.width == 10, "Width should be 10"
assert rect1.height == 5, "Height should be 5"
assert rect1.area() == 50, "Area of 10x5 rectangle should be 50"
assert rect1.perimeter() == 30, "Perimeter of 10x5 rectangle should be 30"

rect2 = Rectangle(7, 3)
assert rect2.area() == 21, "Area of 7x3 rectangle should be 21"
assert rect2.perimeter() == 20, "Perimeter of 7x3 rectangle should be 20"

print("✓ All tests passed!")
```

### Problem 2.2: Add Student Comparison

Add methods to compare students and update grades. Copy your working code from Problem 1.2 and add the new methods.

```python
class Student:
    """
    Enhanced Student with comparison and grade update methods.
    
    New Methods:
        compare_grades(other_student): Returns the name of the student with higher grade
        set_grade(new_grade): Updates grade if valid (0-100), returns status message
    """
    
    def __init__(self, name, grade, student_id):
        # Copy your code from Problem 1.2
        self.name = name
        self.grade = grade
        self.student_id = student_id
    
    def is_passing(self):
        # Copy your code from Problem 1.2
        if self.grade >= 60:
            return True
        else:
            return False
    
    def compare_grades(self, other_student):
        # Return the name of the student with higher grade
        # Your code here
        if self.grade < other_student.grade:
            return other_student.name
        else:
            return self.name
    
    def set_grade(self, new_grade):
        # Check if new_grade is between 0 and 100
        # If valid, update self.grade and return "Grade updated"
        # If invalid, return "Invalid grade" without changing self.grade
        # Your code here
        if new_grade > 100 or new_grade < 0:
            return "Invalid grade"
        self.grade = new_grade
        return "Grade updated"
```

### Problem 2.3: Add Circle Class with Calculations

Create a new `Circle` class with methods to calculate area and circumference.

**Formulas:**
- Area = π * radius²
- Circumference = 2 * π * radius

Use `3.14159` for π

```python
import math

class Circle:
    """
    A class representing a circle.
    
    Attributes:
        radius (float): The circle's radius
    
    Methods:
        area(): Returns the area of the circle
        circumference(): Returns the circumference of the circle
    """
    
    def __init__(self, radius):
        # Your code here
        self.radius = radius
    
    def area(self):
        # Your code here
        return math.pi * (self.radius ** 2)
    
    def circumference(self):
        # Your code here
        return self.radius * 2 * math.pi
```

```python
# Test cases - DO NOT MODIFY
circle1 = Circle(5)
assert abs(circle1.area() - 78.53975) < 0.01, "Area of circle with radius 5 should be ~78.54"
assert abs(circle1.circumference() - 31.4159) < 0.01, "Circumference should be ~31.42"

circle2 = Circle(10)
assert abs(circle2.area() - 314.159) < 0.01, "Area of circle with radius 10 should be ~314.16"

print("✓ All tests passed!")
```

### Problem 3.1: Add Class Attribute and Safe Withdrawal

Enhance the BankAccount class with:
1. A class attribute to track total number of accounts
2. Safe withdrawal that checks for sufficient funds

Copy your code from Problem 2.1 and add these features.

```python
class BankAccount:
    """
    Final BankAccount with class attributes and exception handling.
    
    Class Attributes:
        total_accounts: Tracks the total number of accounts created
    
    Instance Attributes:
        account_number (str): The account number
        owner_name (str): Name of the account owner
        balance (float): The account balance
    
    Methods:
        deposit(amount): Adds amount to balance
        withdraw(amount): Safely subtracts amount if sufficient funds
    """
    
    # Class attribute - initialize to 0
    # Your code here
    total_accounts = 0

    def __init__(self, account_number, owner_name):
        # Copy your previous code
        # Then increment total_accounts
        self.account_number = account_number
        self.owner_name = owner_name
        self.balance = 0
        BankAccount.total_accounts += 1

    def deposit(self, amount):
        # Copy your previous code
        self.balance += amount
    
    def withdraw(self, amount):
        # Check if withdrawal exceeds balance
        # If it does, return "Insufficient funds"
        # Otherwise, subtract amount and return "Withdrawal successful"
        # Your code here
        if amount > self.balance:
            return "Insufficient funds"
        self.balance -= amount
        return "Withdrawal successful"
```

```python
# Test cases - DO NOT MODIFY
# Reset counter for testing
BankAccount.total_accounts = 0

account1 = BankAccount("123", "Alice")
assert BankAccount.total_accounts == 1, "After 1 account, total should be 1"

account2 = BankAccount("456", "Bob")
assert BankAccount.total_accounts == 2, "After 2 accounts, total should be 2"

# Test safe withdrawal
account1.deposit(100)
result1 = account1.withdraw(50)
assert result1 == "Withdrawal successful", "Valid withdrawal should succeed"
assert account1.balance == 50, "Balance should be 50 after withdrawing 50"

result2 = account1.withdraw(100)
assert result2 == "Insufficient funds", "Should return error for insufficient funds"
assert account1.balance == 50, "Balance should remain 50 after failed withdrawal"

print("✓ All tests passed!")
```

### Problem 3.3: Create a Safe Divide Function

Create a function that handles division by zero gracefully using try/except.

```python
def safe_divide(a, b):
    """
    Safely divide two numbers.
    
    Args:
        a (float): Numerator
        b (float): Denominator
    
    Returns:
        float or str: Result of division, or error message if b is 0
    """
    # Use try/except to handle ZeroDivisionError
    # Return "Cannot divide by zero" if error occurs
    # Your code here
    try:
        return a / b
    except ZeroDivisionError as e:
        return "Cannot divide by zero, Only CHuck Norris can divide by zero!"
```

```python
# Test cases - DO NOT MODIFY
assert safe_divide(10, 2) == 5.0, "10 / 2 should equal 5.0"
assert safe_divide(15, 3) == 5.0, "15 / 3 should equal 5.0"
assert safe_divide(10, 0) == "Cannot divide by zero, Only CHuck Norris can divide by zero!", "Division by zero should return error message"
assert safe_divide(0, 5) == 0.0, "0 / 5 should equal 0.0"

print("✓ All tests passed!")
```

### Problem 4.1: Create a Product Class

Create a `Product` class for inventory management with methods to calculate total value and restock items.

```python
class Product:
    """
    A class representing a product in inventory.
    
    Attributes:
        name (str): Product name
        price (float): Price per unit
        quantity (int): Number of units in stock
    
    Methods:
        total_value(): Returns price * quantity (total inventory value)
        restock(amount): Adds amount to quantity
        sell(amount): Reduces quantity by amount if enough stock, returns status message
    """
    
    def __init__(self, name, price, quantity):
        # Your code here
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def total_value(self):
        # Your code here
        return self.price * self.quantity
    
    def restock(self, amount):
        # Your code here
        self.quantity += amount
    
    def sell(self, amount):
        # Check if we have enough quantity
        # If yes, reduce quantity and return "Sale successful"
        # If no, return "Insufficient stock"
        # Your code here
        if amount > self.quantity:
            return "Insufficient stock"
        self.quantity -= amount
        return "Sale successful"
```

```python
# Test cases - DO NOT MODIFY
product1 = Product("Laptop", 999.99, 5)
assert product1.name == "Laptop", "Product name should be 'Laptop'"
assert product1.price == 999.99, "Price should be 999.99"
assert product1.quantity == 5, "Quantity should be 5"
assert abs(product1.total_value() - 4999.95) < 0.01, "Total value should be 4999.95"

# Test restock
product1.restock(3)
assert product1.quantity == 8, "After restocking 3, quantity should be 8"

# Test sell
result1 = product1.sell(5)
assert result1 == "Sale successful", "Should successfully sell 5 items"
assert product1.quantity == 3, "After selling 5, quantity should be 3"

result2 = product1.sell(10)
assert result2 == "Insufficient stock", "Should fail to sell 10 items"
assert product1.quantity == 3, "Quantity should remain 3 after failed sale"

print("✓ All tests passed!")
```

```python
def bayes_theorem(p_b_given_a, p_a, p_b_given_not_a):
    """
    Calculate P(A|B) using Bayes' Theorem.
    
    Args:
        p_b_given_a (float): P(B|A) - likelihood
        p_a (float): P(A) - prior probability
        p_b_given_not_a (float): P(B|not A)
    
    Returns:
        float: P(A|B) - posterior probability
    """
    # Your code here
    # Hint 1: Calculate P(B) using law of total probability
    # P(B) = P(B|A) × P(A) + P(B|not A) × P(not A)
    # Hint 2: Apply Bayes formula: P(A|B) = P(B|A) × P(A) / P(B)
    p_not_a = 1 - p_a
    p_b = (p_b_given_a * p_a) + (p_b_given_not_a * p_not_a)
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b 
```

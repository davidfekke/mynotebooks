## Section 1: Basic Probability

Probability measures how likely an event is to occur, on a scale from 0 (impossible) to 1 (certain).

### Problem 1: Calculate Basic Probability

Create a function that calculates the probability of an event given the number of favorable outcomes and total possible outcomes.

**Formula:** P(Event) = Favorable outcomes / Total outcomes

**What you need to do:**
Write a function that takes two parameters and returns the probability as a decimal.

```python
def calculate_probability(favorable, total):
    """
    Calculate the probability of an event.
    
    Args:
        favorable (int): Number of favorable outcomes
        total (int): Total number of possible outcomes
    
    Returns:
        float: Probability value between 0 and 1
    """
    # Your code here
    return favorable / total
```

```python
# Test cases - DO NOT MODIFY
assert calculate_probability(1, 6) == 1/6, "Rolling a specific number on a die should be 1/6"
assert calculate_probability(1, 2) == 0.5, "Coin flip probability should be 0.5"
assert calculate_probability(3, 6) == 0.5, "Rolling 1, 2, or 3 on a die should be 0.5"
assert calculate_probability(0, 6) == 0, "Impossible event should have probability 0"

print("✓ All tests passed!")
```

## Section 2: Independent Events

Independent events are events where one outcome does not affect the other. For independent events: P(A and B) = P(A) × P(B)

### Problem 2: Multiple Independent Events

Calculate the probability of multiple independent events all occurring.

**What you need to do:**
Write a function that takes a list of probabilities and calculates the probability of all events happening (multiply them together).

```python
def independent_events_probability(probabilities):
    """
    Calculate the probability of multiple independent events all occurring.
    
    Args:
        probabilities (list): List of individual event probabilities
    
    Returns:
        float: Combined probability of all events occurring
    """
    # Your code here
    # Hint: Use a loop to multiply all probabilities together
    if len(probabilities) > 1:
        total_prob = probabilities[0]
        for i in range(1, len(probabilities), 1):
            total_prob *= probabilities[i]
        return total_prob
    else:
        return probabilities[0]
```

```python
# Test cases - DO NOT MODIFY
assert independent_events_probability([0.5, 0.5]) == 0.25, "Two coin flips both heads should be 0.25"
assert independent_events_probability([0.5, 0.5, 0.5]) == 0.125, "Three heads should be 0.125"
assert independent_events_probability([0.7, 0.3, 0.5]) == 0.105, "Mixed probabilities"
assert independent_events_probability([1.0, 1.0]) == 1.0, "Certain events should give 1.0"

print("✓ All tests passed!")
```

## Section 3: Conditional Probability

Conditional probability is the probability of an event occurring given that another event has already occurred. Notation: P(A|B) means "probability of A given B."

### Problem 3: Spam Filter Word Analysis

You're building a simple spam filter. Certain words increase the likelihood that an email is spam.

**What you need to do:**
Write a function that checks if an email contains spam words and returns a spam probability based on what it finds.

```python
def spam_probability(email_words):
    """
    Analyze email words and estimate spam probability.
    
    Rules:
    - If email contains "FREE" or "MONEY": 0.8 probability of spam
    - If email contains "URGENT" or "WINNER": 0.7 probability of spam
    - If email contains multiple spam words: 0.95 probability
    - Otherwise: 0.1 probability (low but not zero)
    
    Args:
        email_words (list): List of words in the email (uppercase)
    
    Returns:
        float: Probability that email is spam (0 to 1)
    """
    # Your code here
    # Hint: Use if/elif/else to check for spam words
    probability_of_spam = 0.1
    probability_of_spam_count = 0
    for word in email_words:
        if word == "FREE" or word == "MONEY":
            probability_of_spam = 0.8
            probability_of_spam_count += 1
        if word == "URGENT" or word == "WINNER":
            probability_of_spam = 0.7
            probability_of_spam_count += 1
    if probability_of_spam_count > 1:
        probability_of_spam = 0.95
    return probability_of_spam
```

```python
# Test cases - DO NOT MODIFY
assert spam_probability(["FREE", "MONEY"]) == 0.95, "Multiple spam words should be 0.95"
assert spam_probability(["FREE", "GIFT"]) == 0.8, "FREE should trigger 0.8"
assert spam_probability(["URGENT", "RESPONSE"]) == 0.7, "URGENT should trigger 0.7"
assert spam_probability(["HELLO", "FRIEND"]) == 0.1, "Normal words should be 0.1"
assert spam_probability(["WINNER", "FREE"]) == 0.95, "Multiple spam words"

print("✓ All tests passed!")
```

## Section 4: Bayes' Theorem

Bayes' Theorem is the foundation of probabilistic AI. It lets us update beliefs based on new evidence.

**Formula:** P(A|B) = [P(B|A) × P(A)] / P(B)

### Problem 4: Bayes' Theorem for Spam Detection

Implement Bayes' Theorem and apply it to the spam filter example from lecture.

**Scenario:** An email contains "FREE MONEY"
- P(Spam) = 0.30
- P("FREE MONEY" | Spam) = 0.80
- P("FREE MONEY" | Not Spam) = 0.02

**What you need to do:**
Calculate P(Spam | "FREE MONEY") using Bayes' Theorem.

```python
def bayes_theorem(p_b_given_a, p_a, p_b_given_not_a):
    """
    Calculate P(A|B) using Bayes' Theorem.
    
    Args:
        p_b_given_a (float): P(B|A) - likelihood
        p_a (float): P(A) - prior probability
        p_b_given_not_a (float): P(B|not A)
    
    Returns:
        float: P(A|B) - posterior probability
    """
    # Your code here
    # Hint 1: Calculate P(B) using law of total probability
    # P(B) = P(B|A) × P(A) + P(B|not A) × P(not A)
    # Hint 2: Apply Bayes formula: P(A|B) = P(B|A) × P(A) / P(B)
    p_not_a = 1 - p_a
    p_b = (p_b_given_a * p_a) + (p_b_given_not_a * p_not_a)
    if p_b == 0:
        return 0.0
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b 
```

```python
# Test cases - DO NOT MODIFY
# Spam filter example from lecture
spam_prob = bayes_theorem(p_b_given_a=0.80, p_a=0.30, p_b_given_not_a=0.02)
assert abs(spam_prob - 0.944) < 0.01, "Spam probability should be about 92.3%"

# Medical test example
disease_prob = bayes_theorem(p_b_given_a=0.99, p_a=0.001, p_b_given_not_a=0.01)
assert abs(disease_prob - 0.09) < 0.01, "Disease probability should be about 9%"

print("✓ All tests passed!")
```

## Section 5: Normal Distribution

The normal distribution (bell curve) is everywhere in statistics and machine learning. It's defined by two parameters: mean (center) and standard deviation (spread).

### Problem 5: Calculate Mean and Standard Deviation

Calculate the mean and standard deviation of a dataset. This is a refresher from last week's statistics primer.

**What you need to do:**
1. Calculate mean: sum of all values / count
2. Calculate standard deviation: sqrt of average squared differences from mean

```python
def calculate_mean(numbers):
    return (sum(numbers) / len(numbers))

def calculate_standard_deviation(numbers):
    mean = calculate_mean(numbers)
    deviation_from_mean = [n - mean for n in numbers]
    sum_of_deviation_squared = sum([n ** 2 for n in deviation_from_mean]) 
    variance = sum_of_deviation_squared / (len(numbers) - 1)
    return variance ** 0.5

def calculate_mean_and_std(data):
    """
    Calculate mean and standard deviation of a dataset.
    
    Args:
        data (list): List of numbers
    
    Returns:
        tuple: (mean, standard_deviation)
    """
    # Your code here
    # Hint 1: Use a loop to calculate sum, then divide by length for mean
    # Hint 2: For std dev, calculate squared differences from mean, average them, then sqrt
    return (calculate_mean(data), calculate_standard_deviation(data))
```

```python
# Test cases - DO NOT MODIFY
mean1, std1 = calculate_mean_and_std([1, 2, 3, 4, 5])
assert mean1 == 3.0, "Mean of 1-5 should be 3"
assert abs(std1 - 1.581) < 0.01, "Std dev should be about 1.414"

mean2, std2 = calculate_mean_and_std([10, 20, 30, 40, 50])
assert mean2 == 30.0, "Mean should be 30"
assert abs(std2 - 15.81) < 0.1, "Std dev should be about 14.14"

mean3, std3 = calculate_mean_and_std([5, 5, 5, 5, 5])
assert mean3 == 5.0, "Mean of all 5s should be 5"
assert std3 == 0.0, "Std dev of identical values should be 0"

print("✓ All tests passed!")
```

## Section 6: Sample vs Population

In machine learning, we never have access to all possible data (population). We only have a sample (training data). Understanding this relationship is crucial for model generalization.

### Problem 6: Calculate Sample Statistics

Take a random sample from a population and calculate its mean. Compare it to the true population mean.

**What you need to do:**
1. Take a random sample from the population
2. Calculate the sample mean
3. Calculate the population mean
4. Return both values

```python
import random
import statistics as st

def sample_vs_population_mean(population, sample_size):
    """
    Compare sample mean to population mean.
    
    Args:
        population (list): The full population of values
        sample_size (int): Number of items to sample
    
    Returns:
        tuple: (sample_mean, population_mean, difference)
    """
    # Your code here
    # Hint 1: Use random.sample(population, sample_size) to get random sample
    # Hint 2: Calculate mean for both sample and population
    # Hint 3: Calculate absolute difference
    random_sample = random.sample(population, sample_size)
    sample_mean = st.mean(random_sample)
    population_mean = st.mean(population)
    if population_mean > sample_mean:
        diff = population_mean - sample_mean
    else:
        diff = sample_mean - population_mean
    return (sample_mean, population_mean, diff)
```

```python
# Test cases - DO NOT MODIFY
test_population = list(range(1, 101))  # Population: 1 to 100
sample_mean, pop_mean, diff = sample_vs_population_mean(test_population, 20)

assert pop_mean == 50.5, "Population mean of 1-100 should be 50.5"
assert 30 <= sample_mean <= 70, "Sample mean should be reasonable"
assert diff >= 0, "Difference should be non-negative"
assert diff == abs(sample_mean - pop_mean), "Difference calculation"

print("✓ All tests passed!")
```

```python
import statistics as st

def calculate_standard_dev(numbers):
    mean = sum(numbers) / len(numbers)
    print(f"Mean: {mean}")
    deviation_from_mean = [n - mean for n in numbers]
    print(f"Deviation_from_mean: {deviation_from_mean}")
    sum_of_deviation_squared = sum([n ** 2 for n in deviation_from_mean]) 
    print(f"Sum of deviation squared: {sum_of_deviation_squared}")
    variance = sum_of_deviation_squared / (len(numbers) - 1)
    print(f"Variance: {variance}")
    print(f"Square root of Variance: {variance ** 0.5}")
    return variance ** 0.5

test_data = [72, 68, 75, 70, 73, 69, 71]
standard_dev = calculate_standard_dev(test_data)
assert standard_dev == st.stdev(test_data), "Deviation is incorrect value"

print("Test passed!")
standard_dev
```

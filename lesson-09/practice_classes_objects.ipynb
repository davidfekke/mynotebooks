```python
class Movie:
    # This is a static property
    total_movies = 0

    def __init__(self, title, director, genre):
        self.title = title
        self.director = director
        self.runtime = genre
        Movie.total_movies += 1
    
    def get_info(self):
        return f"{self.title} by {self.director}"
    
    def is_long_movie(self, runtime):
        if runtime > 150:
            return "This is a long movie!"
        return "Standard length!"
    
```

```python
die_hard = Movie("Die Hard", "John McTiernan", "Action")

die_hard.title = "Die Harder"
print(die_hard.title)

movie1 = Movie("Inception", "Christopher Nolan", "Sci-Fi")
movie2 = Movie("The Matrix", "Wachowskis", "Sci-Fi")

print(movie2.get_info())
print(movie2.is_long_movie(151))
print(die_hard.total_movies)
```

```python
def safe_divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        return "Cannot divide by zero!"
    except TypeError:
        return "Please provide numbers only!"
    

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide(10, "a"))
```

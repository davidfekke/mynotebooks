## Part 1: Data Preprocessing

**The Problem:** ShopSmart's database is old and glitchy. Product names come in with weird spacing and capitalization, and prices are stored as strings with '$' symbols. We can't do any analysis until this data is clean.

**Your Task:** Create a function that takes messy lists and returns clean, usable data.

### Example Verification
**Input:**
- `names = ["  Apple ", "BANANA"]`
- `prices = ["$1.50", "$0.75"]`

**Expected Output:**
- `(['apple', 'banana'], [1.5, 0.75])`

````python
def clean_product_data(product_names, product_prices):
    """
    Clean product names and prices.
    
    Args:
        product_names (list): List of messy product names
        product_prices (list): List of price strings
        
    Returns:
        tuple: (cleaned_names, cleaned_prices)
    """
    cleansed_product_names = [pn.strip().lower() for pn in product_names]
    cleansed_product_prices = [float(pp.replace("$", "")) for pp in product_prices]
    return (cleansed_product_names, cleansed_product_prices)
````

````python
# Basic Verification - Add your own tests too!
names_test = ["  Apple ", "BANANA"]
prices_test = ["$1.50", "$0.75"]
clean_names, clean_prices = clean_product_data(names_test, prices_test)

assert clean_names == ["apple", "banana"], "Names not cleaned correctly"
assert clean_prices == [1.50, 0.75], "Prices not cleaned correctly"
print("[PASS] Part 1 Basic Tests Passed!")
````

## Part 2: Sales Metrics & Analysis

**The Problem:** The client needs to know their total revenue and the average (mean) transaction value to report to their investors. They also want to flag any "high value" transactions (over $100) for special rewards. We also need to understand the spread of the data.

**Your Task:** Analyze a list of transaction amounts to calculate these metrics and return them in a dictionary.
*Hint: You might want to use `import statistics` for some of these calculations.*

**Required Keys:**
- `'total_revenue'`
- `'mean'`
- `'median'`
- `'mode'`
- `'stdev'` (Standard Deviation)
- `'25th_percentile'`
- `'high_value_count'`

### Example Verification
**Input:** `sales = [50.0, 150.0, 20.0, 200.0, 80.0, 20.0]`

**Expected Output:**
```python
{
    'total_revenue': 520.0,
    'mean': 86.67, # approx
    'median': 65.0,
    'mode': 20.0,
    'stdev': 74.74, # approx
    '25th_percentile': 27.5, # approx
    'high_value_count': 2
}
```

````python
import statistics as st

def analyze_revenue(transactions):
    """
    Analyze sales data to find total, mean, median, mode, stdev, percentile, and high-value counts.
    
    Args:
        transactions (list): List of floats representing transaction amounts
        
    Returns:
        dict: Dictionary with all required keys
    """
    q1 = st.quantiles(transactions, n=4)[0]
    q2 = st.quantiles(transactions, n=100)[24]
    print(q1 == q2)
    return {
        "total_revenue": sum([t for t in transactions]), 
        "mean": st.mean(transactions), 
        "median": st.median(transactions), 
        "mode": st.mode(transactions), 
        "stdev": st.stdev(transactions), 
        "25_percentile": st.quantiles(transactions, n=4)[0], 
        "high_value_count": len([c for c in transactions if c > 100])
    }
````

````python
# Basic Verification - Add your own tests too!
sales_test = [50.0, 150.0, 20.0, 200.0, 80.0, 20.0]
results = analyze_revenue(sales_test)

assert results['total_revenue'] == 520.0, "Total revenue incorrect"
assert abs(results['mean'] - 86.666) < 0.01, "Mean value incorrect"
assert results['median'] == 65.0, "Median value incorrect"
assert results['mode'] == 20.0, "Mode value incorrect"
assert abs(results['stdev'] - 73.66) < 0.1, "Standard Deviation incorrect"
assert results['high_value_count'] == 2, "High value count incorrect"
print("[PASS] Part 2 Basic Tests Passed!")
````

## Part 3: Pricing & Taxation Engine

**The Problem:** We are expanding to new regions with different tax rates. We need a flexible tool that can calculate the final price of an item, optionally applying a discount for our members.

**Your Task:** Create a function that handles these calculations dynamically.
1.  **Tax Logic:**
    - If category is `'grocery'`, tax is **0%**.
    - If category is `'electronics'`, tax is **10%**.
    - If category is `'clothing'`, tax is **5%**.
    - For any other category, tax is **8%**.
2.  **Discount Logic:**
    - If `is_member` is **True**, apply a **10% discount** to the price *before* tax.

### Example Verification
| Call | Expected Output | Logic |
| :--- | :--- | :--- |
| `calculate_price(100.0, 'grocery')` | `100.0` | 0% tax |
| `calculate_price(100.0, 'electronics')` | `110.0` | 10% tax |
| `calculate_price(100.0, 'other', is_member=True)` | `97.20` | (100 * 0.90) * 1.08 |

````python
def calculate_price(price, category, is_member=False):
    """
    Calculate the final price based on category tax and optional member discount.
    
    Args:
        price (float): Base price
        category (str): Product category ('grocery', 'electronics', 'clothing', or other)
        is_member (bool): If True, apply 10% discount before tax
        
    Returns:
        float: Final price rounded to 2 decimal places
    """
    total_price = price
    if is_member:
        total_price *= 0.9
    
    if category == "grocery":
        return round(total_price, 2)
    elif category == "electronics":
        return round(total_price * 1.10, 2)
    elif category == "clothing":
        return round(total_price * 1.05, 2)
    else:
        return round(total_price * 1.08, 2)
    
````

````python
# Basic Verification - Add your own tests too!
assert calculate_price(100.0, 'grocery') == 100.0, "Grocery tax failed"
assert calculate_price(100.0, 'electronics') == 110.0, "Electronics tax failed"
assert calculate_price(100.0, 'clothing') == 105.0, "Clothing tax failed"
assert calculate_price(100.0, 'other', is_member=True) == 97.20, "Member discount logic failed"
print("[PASS] Part 3 Basic Tests Passed!")
````

## Part 4: Customer Segmentation

**The Problem:** Our data team has given us a raw list of transaction dictionaries. We need to answer two questions: 
1. How many *unique* customers shopped with us?
2. How much revenue did we make in each category (Electronics vs Clothing)?

**Your Task:** Process the list of dictionaries to extract these insights.
*Hint: Think about using a **Set**. In Python, converting a list to a set (`set(my_list)`) automatically removes all duplicate values, leaving you with only the unique items. This is much faster than checking one by one!*

### Example Verification
**Input:**
```python
data = [
    {'id': 101, 'category': 'Electronics', 'amount': 200.0},
    {'id': 102, 'category': 'Clothing', 'amount': 50.0},
    {'id': 101, 'category': 'Electronics', 'amount': 100.0}
]
```

**Expected Output:**
- Unique Customers: `2` (IDs 101 and 102)
- Category Totals: `{'Electronics': 300.0, 'Clothing': 50.0}`

````python
def segment_customers(transactions):
    """
    Analyze transaction dictionaries.
    
    Args:
        transactions (list): List of dicts
                             
    Returns:
        tuple: (unique_customer_count, category_totals_dict)
    """
    unique_categories = set([c["category"] for c in transactions])
    cat_totals = {}
    for cat in unique_categories:
        cat_totals[cat] = sum([a["amount"] for a in transactions if a["category"] == cat])
    return (len(set([t["id"] for t in transactions])), cat_totals)
````

````python
# Basic Verification - Add your own tests too!
data_test = [
    {'id': 101, 'category': 'Electronics', 'amount': 200.0},
    {'id': 102, 'category': 'Clothing', 'amount': 50.0},
    {'id': 101, 'category': 'Electronics', 'amount': 100.0}
]
unique_count, cat_totals = segment_customers(data_test)

assert unique_count == 2, "Unique customer count incorrect"
assert cat_totals['Electronics'] == 300.0, "Electronics total incorrect"
print("[PASS] Part 4 Basic Tests Passed!")
````

## Part 5: The Transaction System Upgrade

**The Problem:** As ShopSmart grows, using simple dictionaries for transactions is becoming messy. We need a more robust way to handle this data. We're upgrading to an Object-Oriented approach.

**Your Task:** Create a `Transaction` class that can store data and perform actions on itself.

### Example Verification
```python
# 1. Create a transaction
t = Transaction(id=1, amount=200.0, category="Electronics")

# 2. Check if it's high value
print(t.is_high_value()) # Should print True

# 3. Apply a discount
t.apply_discount(0.10) # Apply 10% discount
print(t.amount) # Should print 180.0
```

````python
class Transaction:
    def __init__(self, t_id, amount, category):
        """
        Initialize the Transaction.
        """
        self.t_id = t_id
        self.amount = amount
        self.category = category
        
    def is_high_value(self, threshold=100.0):
        """
        Check if transaction is high value.
        Returns True if amount > threshold.
        """
        if self.amount > threshold:
            return True
        return False
        
    def apply_discount(self, percent):
        """
        Apply a percentage discount to the amount.
        Updates self.amount.
        """
        self.amount *= (1 - percent)
````

````python
# Basic Verification - Add your own tests too!
t = Transaction(1, 200.0, "Electronics")
assert t.is_high_value() == True, "High value check failed"
t.apply_discount(0.10)
assert t.amount == 180.0, "Discount application failed"
print("[PASS] Part 5 Basic Tests Passed!")
````

## Part 6: Future Forecasting (Advanced)

**The Problem:** Finally, we need to look ahead. Use your math skills to help us predict trends and rank products.

**Your Task:**
1.  **Growth Rate:** Calculate the slope (rate of change) between two months.
2.  **Product Score:** Calculate a weighted score using a dot product.

### Example Verification
**Growth Rate:**
- Month 1: `1000`, Month 2: `1500` -> Slope: `500.0`

**Product Score:**
- Features: `[0.8, 0.9, 0.5]`, Weights: `[0.2, 0.5, 0.3]`
- Calculation: `(0.8*0.2) + (0.9*0.5) + (0.5*0.3)` -> Score: `0.76`

````python
def calculate_growth_rate(month1_sales, month2_sales):
    """
    Calculate the rate of change (slope) between two months.
    """
    # Your code here
    return month2_sales - month1_sales

def calculate_product_score(features, weights):
    """
    Calculate product score using dot product (weighted sum).
    """
    # Your code here
    score = 0
    for i in range(len(features)):
        score += features[i] * weights[i]
    return score
````

````python
# Basic Verification - Add your own tests too!
assert calculate_growth_rate(1000, 1500) == 500.0, "Growth rate incorrect"
assert abs(calculate_product_score([0.8, 0.9, 0.5], [0.2, 0.5, 0.3]) - 0.76) < 0.001, "Product score incorrect"
print("[PASS] Part 6 Basic Tests Passed!")
````

### Problem 1.1: Vector Addition

In AI, we often need to combine vectors (like adding two sets of features together).

**What you need to do:**
Create a function that adds two vectors element-wise. For example, [1, 2, 3] + [4, 5, 6] = [5, 7, 9].

```python
import numpy as np

def add_vectors(vector1, vector2):
    """
    Add two vectors element-wise.
    
    Args:
        vector1 (list): First vector
        vector2 (list): Second vector
    
    Returns:
        list: Resulting vector with element-wise sums
    """
    # Your code here
    if len(vector1) != len(vector2):
        return "Vectors must be the same length!"
    result = []
    for i in range(len(vector1)):
        result.append(vector1[i] + vector2[i])
    return result

def add_vectors_with_numpy(vector1, vector2):
    result = np.add(vector1, vector2)
    return list(result) 

```

```python
# Test cases - DO NOT MODIFY
assert add_vectors([1, 2, 3], [4, 5, 6]) == [5, 7, 9], "Adding [1,2,3] + [4,5,6] should give [5,7,9]"
assert add_vectors([0, 0], [1, 1]) == [1, 1], "Adding zero vector should return the other vector"
assert add_vectors([10, 20, 30], [5, 10, 15]) == [15, 30, 45], "Should work with larger numbers"

assert add_vectors_with_numpy([1, 2, 3], [4, 5, 6]) == [5, 7, 9], "Adding [1,2,3] + [4,5,6] should give [5,7,9]"

print("✓ All tests passed!")
```

### Problem 1.2: Dot Product

The dot product is one of the most important operations in AI! It measures how "aligned" two vectors are.

**What you need to do:**
Multiply corresponding elements of two vectors and sum the results. For example:
- [1, 2, 3] · [4, 5, 6] = (1×4) + (2×5) + (3×6) = 4 + 10 + 18 = 32

```python
import numpy as np

def dot_product(vector1, vector2):
    """
    Compute the dot product of two vectors.
    
    Args:
        vector1 (list): First vector
        vector2 (list): Second vector
    
    Returns:
        float: The dot product (sum of element-wise products)
    """
    # Your code here
    if len(vector1) != len(vector2):
        return "Vectors must be the same length!"
    intermediate_result = []
    for i in range(len(vector1)):
        intermediate_result.append(vector1[i] * vector2[i])
    return sum(intermediate_result)

def dot_product_with_numpy(vector1, vector2):
    return np.dot(vector1, vector2)
```

```python
# Test cases - DO NOT MODIFY
assert dot_product([1, 2, 3], [4, 5, 6]) == 32, "Dot product should be 32"
assert dot_product([1, 0], [0, 1]) == 0, "Perpendicular vectors have dot product 0"
assert dot_product([2, 3], [2, 3]) == 13, "Dot product with itself"

assert dot_product_with_numpy([1, 2, 3], [4, 5, 6]) == 32, "Dot product should be 32"
assert dot_product_with_numpy([1, 0], [0, 1]) == 0, "Perpendicular vectors have dot product 0"
assert dot_product_with_numpy([2, 3], [2, 3]) == 13, "Dot product with itself"

print("✓ All tests passed!")
```

### Problem 1.3: Create Matrix from Rows

Matrices organize data in rows and columns. In AI, your entire dataset is just one big matrix!

**What you need to do:**
Combine individual lists (rows) into a 2D matrix (list of lists).

```python
def create_matrix(*rows):
    """
    Create a matrix from individual row vectors.
    
    Args:
        *rows: Variable number of lists representing rows
    
    Returns:
        list: A 2D list (matrix) containing all rows
    """
    # Your code here
    result = []
    for row in rows:
        result.append(row)
    return result
```

```python
# Test cases - DO NOT MODIFY
assert create_matrix([1, 2], [3, 4]) == [[1, 2], [3, 4]], "2x2 matrix creation"
assert create_matrix([1, 2, 3], [4, 5, 6], [7, 8, 9]) == [[1, 2, 3], [4, 5, 6], [7, 8, 9]], "3x3 matrix creation"
assert create_matrix([10]) == [[10]], "1x1 matrix"

print("✓ All tests passed!")
```

### Problem 1.4: Get Matrix Dimensions

Before working with a matrix, we need to know its size!

**What you need to do:**
Return the dimensions of a matrix as (number_of_rows, number_of_columns).

```python
def matrix_dimensions(matrix):
    """
    Get the dimensions of a matrix.
    
    Args:
        matrix (list): A 2D list representing a matrix
    
    Returns:
        tuple: (rows, columns)
    """
    # Your code here
    width = len(matrix)
    height = len(matrix[0])
    return width, height
```

```python
# Test cases - DO NOT MODIFY
assert matrix_dimensions([[1, 2], [3, 4]]) == (2, 2), "2x2 matrix"
assert matrix_dimensions([[1, 2, 3], [4, 5, 6]]) == (2, 3), "2x3 matrix"
assert matrix_dimensions([[1], [2], [3]]) == (3, 1), "3x1 matrix"

print("✓ All tests passed!")
```

### Problem 1.5: Matrix Element Access

Access specific elements in a matrix by row and column position.

**What you need to do:**
Given a matrix and a row/column index, return the element at that position.

```python
def get_element(matrix, row, col):
    """
    Get the element at a specific row and column in a matrix.
    
    Args:
        matrix (list): A 2D list representing a matrix
        row (int): Row index (0-based)
        col (int): Column index (0-based)
    
    Returns:
        The element at the specified position
    """
    # Your code here
    return matrix[row][col]
```

```python
# Test cases - DO NOT MODIFY
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
assert get_element(matrix, 0, 0) == 1, "Top-left element"
assert get_element(matrix, 1, 1) == 5, "Center element"
assert get_element(matrix, 2, 2) == 9, "Bottom-right element"
assert get_element(matrix, 0, 2) == 3, "Top-right element"

print("✓ All tests passed!")
```

### Problem 2.1: Calculate Slope (Rate of Change)

A derivative measures how fast something is changing. The slope between two points is a simple example!

**What you need to do:**
Calculate the slope (rate of change) between two points using the formula: (y2 - y1) / (x2 - x1).

```python
def calculate_slope(x1, y1, x2, y2):
    """
    Calculate the slope (rate of change) between two points.
    
    Args:
        x1 (float): x-coordinate of first point
        y1 (float): y-coordinate of first point
        x2 (float): x-coordinate of second point
        y2 (float): y-coordinate of second point
    
    Returns:
        float: The slope (rate of change)
    """
    # Your code here
    return (y2 - y1) / (x2 - x1)
```

```python
# Test cases - DO NOT MODIFY
assert calculate_slope(0, 0, 1, 2) == 2.0, "Slope should be 2"
assert calculate_slope(0, 0, 2, 2) == 1.0, "Slope should be 1"
assert calculate_slope(1, 3, 3, 7) == 2.0, "Slope should be 2"

print("✓ All tests passed!")
```

### Problem 2.2: Find Direction to Minimize

AI uses derivatives to figure out which direction reduces error!

**What you need to do:**
Given a derivative value:
- If derivative > 0: we should DECREASE input to minimize (return "decrease")
- If derivative < 0: we should INCREASE input to minimize (return "increase")
- If derivative = 0: we're at a minimum! (return "at minimum")

```python
def find_direction(derivative):
    """
    Determine which direction to move to minimize a function.
    
    Args:
        derivative (float): The derivative value at current point
    
    Returns:
        str: "increase", "decrease", or "at minimum"
    """
    # Your code here
    if derivative > 0:
        return "decrease"
    elif derivative < 0:
        return "increase"
    else:
        return "at minimum"
```

```python
# Test cases - DO NOT MODIFY
assert find_direction(5) == "decrease", "Positive derivative means decrease input"
assert find_direction(-3) == "increase", "Negative derivative means increase input"
assert find_direction(0) == "at minimum", "Zero derivative means at minimum"

print("✓ All tests passed!")
```

### Problem 2.3: Apply Chain Rule

Neural networks have multiple layers. The chain rule helps us track how changes flow through all layers!

**What you need to do:**
Given a list of derivatives (one from each layer), multiply them all together. This is the chain rule in action!

```python
def apply_chain_rule(derivatives):
    """
    Apply the chain rule by multiplying derivatives from each layer.
    
    Args:
        derivatives (list): List of derivative values from each layer
    
    Returns:
        float: The product of all derivatives
    """
    # Your code here
    if len(derivatives) < 1:
        return 0
    acc = derivatives[0]
    for i in range(1, len(derivatives), 1):
        acc *= derivatives[i]
    return acc 
```

```python
# Test cases - DO NOT MODIFY
assert apply_chain_rule([2, 3]) == 6, "2 × 3 = 6"
assert apply_chain_rule([2, 3, 4]) == 24, "2 × 3 × 4 = 24"
assert apply_chain_rule([0.5, 0.5, 0.5]) == 0.125, "0.5 × 0.5 × 0.5 = 0.125"

print("✓ All tests passed!")
```

### Problem 2.4: Compute Gradient Vector

A gradient is a vector of partial derivatives - one for each parameter in your model!

**What you need to do:**
Take individual partial derivative values and combine them into a gradient vector (list).

```python
def compute_gradient(*partial_derivatives):
    """
    Combine partial derivatives into a gradient vector.
    
    Args:
        *partial_derivatives: Variable number of partial derivative values
    
    Returns:
        list: The gradient vector
    """
    # Your code here
    result = []
    for item in partial_derivatives:
        result.append(item)
    return result
```

```python
# Test cases - DO NOT MODIFY
assert compute_gradient(2, 3) == [2, 3], "2D gradient"
assert compute_gradient(1, 2, 3) == [1, 2, 3], "3D gradient"
assert compute_gradient(0.5, -0.3, 0.8) == [0.5, -0.3, 0.8], "Gradient with decimals"

print("✓ All tests passed!")
```

**What you need to do:**

```python
import numpy as np

def find_most_impactful_parameter(gradient):
    """
    Find which parameter has the largest gradient (biggest impact).
    
    Args:
        gradient (list): The gradient vector
    
    Returns:
        int: Index of the parameter with largest absolute gradient value
    """
    # Your code here
    find_most_impactful_parameter_with_numpy(gradient)
    
    differences = []
    for i in range(len(gradient) - 1):
        first = gradient[i]
        second = gradient[i+1]
        if first > second:
            current_diff = first - second
        else:
            current_diff = second - first
        differences.append(current_diff)
    return differences.index(max(differences)) + 1
    
def find_most_impactful_parameter_with_numpy(gradient):
    x = np.array(gradient)
    grads = [i ** 2 for i in np.diff(x)]

    print(grads.index(max(grads)) + 1)
```

```python
def calculate_mode(numbers):
    """
    Calculate the mode (most frequently occurring value) of a list of numbers.
    
    Args:
        numbers (list): List of numbers
    
    Returns:
        The mode of the list
    """
    # Your code here
    frequency = {}
    for number in numbers:
        if number in frequency:
            frequency[number] += 1
        else:
            frequency[number] = 1
    most_frequent = max(frequency, key=frequency.get)
    max_value = max(frequency.values())
    print(max_value)
    return most_frequent
```

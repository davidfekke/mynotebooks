### Problem 1.1: BMI Calculator

Body Mass Index (BMI) is calculated using the formula:

**BMI = weight (kg) / height² (m²)**

Create a function that calculates BMI given weight in kilograms and height in meters.

````python
def calculate_bmi(weight_kg, height_m):
    """
    Calculate Body Mass Index.
    
    Args:
        weight_kg (float): Weight in kilograms
        height_m (float): Height in meters
    
    Returns:
        float: BMI value rounded to 1 decimal place
    """
    # Your code here
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 1)
````

````python
# Test cases - DO NOT MODIFY
assert calculate_bmi(70, 1.75) == 22.9, "BMI should be 22.9"
assert calculate_bmi(85, 1.80) == 26.2, "BMI should be 26.2"
assert calculate_bmi(60, 1.65) == 22.0, "BMI should be 22.0"

print("✓ All tests passed!")
````

### Problem 1.2: Temperature Filter

Given a list of daily temperatures in Celsius, filter out only the days where the temperature was above freezing (0°C).

Return a new list containing only temperatures greater than 0.

````python
def filter_above_freezing(temperatures):
    """
    Filter temperatures to only those above freezing (0°C).
    
    Args:
        temperatures (list): List of temperature values in Celsius
    
    Returns:
        list: Only temperatures > 0
    """
    # Your code here
    positive_temps = []
    for temp in temperatures:
        if temp > 0:
            positive_temps.append(temp)
    # return positive_temps
    return [t for t in temperatures if t > 0]

# output = [t for t in temperatures if t > 0]
````

````python
# Test cases - DO NOT MODIFY
assert filter_above_freezing([-5, 0, 3, -2, 10, 15]) == [3, 10, 15], "Should return [3, 10, 15]"
assert filter_above_freezing([-10, -5, -1]) == [], "All below/at freezing, return empty list"
assert filter_above_freezing([5, 10, 15, 20]) == [5, 10, 15, 20], "All above freezing"
assert filter_above_freezing([0, 0, 1, 0]) == [1], "0 is not above freezing"

print("✓ All tests passed!")
````

### Problem 1.4: Random Breakout Room Generator

Create a function that randomly assigns people to breakout rooms - just like Zoom did to create your current groups!

Given a list of names and a room size, randomly shuffle the names and group them into rooms.

**Hints:** 
- Use Python's `random` library and `random.shuffle()` to randomize the order
- Make a **copy** of the names list first (use `.copy()`) so you don't modify the original
- Use `range(start, stop, step)` with `room_size` as the step to create room groups
- You can slice a list like `names[0:3]` to get the first 3 elements

**Note:** If the names don't divide evenly, the last room can be smaller.

**Example approach:**
1. Make a copy of the names list
2. Shuffle the copy
3. Loop through the shuffled list in chunks of `room_size`
4. Create a room for each chunk

````python
import random
import math

def create_breakout_rooms(names, room_size):
    """
    Randomly assign people to breakout rooms.
    
    Args:
        names (list): List of participant names
        room_size (int): Maximum number of people per room
    
    Returns:
        list: List of lists, where each inner list is a room with assigned names
    
    Example:
        Input: ['Alice', 'Bob', 'Carol', 'David', 'Eve'], room_size=2
        Output: [['Carol', 'Eve'], ['David', 'Alice'], ['Bob']] (order will vary)
    """
    # Your code here
    parent_list = []
    shuffled = names.copy()
    random.shuffle(shuffled)
   
    for index in range(0, len(shuffled), room_size):
        parent_list.append(shuffled[index:index+room_size])
        
    return parent_list

# result1 = create_breakout_rooms(['A', 'B', 'C', 'D', 'E', 'F'], 3)

````

````python
# Test cases - DO NOT MODIFY
result1 = create_breakout_rooms(['A', 'B', 'C', 'D', 'E', 'F'], 3)
assert len(result1) == 2, "Should create 2 rooms"
assert len(result1[0]) == 3 and len(result1[1]) == 3, "Each room should have 3 people"
assert sorted([name for room in result1 for name in room]) == ['A', 'B', 'C', 'D', 'E', 'F'], "All names should be present"

result2 = create_breakout_rooms(['A', 'B', 'C', 'D', 'E'], 2)
assert len(result2) == 3, "Should create 3 rooms"
assert len(result2[-1]) <= 2, "Last room can be smaller"

print("✓ All tests passed!")
print(f"\nExample: {result1}")
print("Fun fact: You just built the algorithm that Zoom uses to assign breakout rooms!")
````

### Problem 2: Temperature Data Analyzer

You're analyzing a week's worth of temperature data. Given a list of daily temperatures (in Fahrenheit), calculate statistics and classify the week.

**Requirements:**

1. Calculate the **mean (average)** temperature
2. Calculate the **median** temperature
3. Count the number of **extreme days**:
   - Very hot: temperature > 90°F
   - Freezing: temperature < 32°F
4. Classify the week based on the mean:
   - "Hot" if mean > 80
   - "Cold" if mean < 50
   - "Moderate" otherwise
5. Return a dictionary with all these statistics

**Input format:** Simple list of numbers (temperatures in Fahrenheit)
```python
temperatures = [72, 68, 75, 70, 73, 69, 71]
```

**Reminder:** 
- Mean = sum of all values / count of values
- Median = middle value when sorted (or average of two middle values if even count)

````python
import statistics

def analyze_temperature_data(temperatures):
    """
    Analyze temperature data and return statistics.
    
    Args:
        temperatures (list): List of daily temperatures in Fahrenheit
    
    Returns:
        dict: Dictionary with keys 'mean', 'median', 'extreme_days', 'classification'
    """
    mean = statistics.mean(temperatures)
    median = statistics.median(temperatures)
    extreme_days = len(list(filter(lambda t: t > 90 or t < 32, temperatures)))
    extreme_days2 = len([t for t in temperatures if t > 90 or t < 32])
    classification = "Moderate"

    if mean > 80:
        classification = "Hot"
    elif mean < 50:
        classification = "Cold"
    # else:
    #     classification = "Moderate"
    return {
        'mean': mean, 
        'median': median, 
        'extreme_days': extreme_days2, 
        'classification': classification
    }
    
````

````python
# Test cases - DO NOT MODIFY
temps1 = [72, 68, 75, 70, 73, 69, 71]
# Mean: (72+68+75+70+73+69+71) / 7 = 498 / 7 = 71.14
# Median: sorted = [68, 69, 70, 71, 72, 73, 75], middle = 71
# Extreme days: 0 (none > 90 or < 32)
# Classification: "Moderate" (mean is 71.14, between 50 and 80)
result1 = analyze_temperature_data(temps1)
assert round(result1['mean'], 2) == 71.14, "Mean should be 71.14"
assert result1['median'] == 71, "Median should be 71"
assert result1['extreme_days'] == 0, "Should have 0 extreme days"
assert result1['classification'] == "Moderate", "Should be classified as Moderate"

temps2 = [95, 92, 88, 91, 94, 89, 93]
# Mean: (95+92+88+91+94+89+93) / 7 = 642 / 7 = 91.71
# Median: sorted = [88, 89, 91, 92, 93, 94, 95], middle = 92
# Extreme days: 5 (95, 92, 91, 94, 93 are all > 90)
# Classification: "Hot" (mean > 80)
result2 = analyze_temperature_data(temps2)
assert round(result2['mean'], 2) == 91.71, "Mean should be 91.71"
assert result2['median'] == 92, "Median should be 92"
assert result2['extreme_days'] == 5, "Should have 5 extreme days"
assert result2['classification'] == "Hot", "Should be classified as Hot"

temps3 = [28, 35, 40, 30, 33, 38]
# Mean: (28+35+40+30+33+38) / 6 = 204 / 6 = 34.0
# Median: sorted = [28, 30, 33, 35, 38, 40], middle = (33+35)/2 = 34.0
# Extreme days: 2 (28 and 30 are < 32)
# Classification: "Cold" (mean < 50)
result3 = analyze_temperature_data(temps3)
assert result3['mean'] == 34.0, "Mean should be 34.0"
assert result3['median'] == 34.0, "Median should be 34.0"
assert result3['extreme_days'] == 2, "Should have 2 extreme days"
assert result3['classification'] == "Cold", "Should be classified as Cold"

print("✓ All tests passed!")
````

### Utility Functions

Here is a utility function that might be helpful:

````python
def calculate_item_total(item):
    """
    Calculate total for an item with quantity discount.
    If quantity >= 3, apply 10% discount to that item.
    
    Args:
        item (dict): Item with 'price' and 'quantity' keys
    
    Returns:
        float: Total price for this item
    """
    base_total = item['price'] * item['quantity']
    if item['quantity'] >= 3:
        return base_total * 0.90  # 10% discount
    return base_total
````

### Problem 3: Online Shopping Cart Calculator

You're building a shopping cart total calculator for a small online store.

**Given data:**
- A shopping cart (list of items, each with item name, price, quantity, and category)
- Whether the customer is a member

**Your function should calculate the final total including:**

1. **Quantity discounts:** If buying 3 or more of the same item, that item gets 10% off
2. **Membership discount:** Members get 15% off the entire order
3. **Shipping:** Orders over $100 get free shipping, otherwise add $10 for shipping
4. **Tax:** Apply 8% sales tax on the discounted total (before shipping)
5. **Return the final total** (after all discounts, tax, and shipping)

**Order of operations:**
- Calculate item totals (with quantity discounts)
- Apply membership discount to subtotal
- Calculate tax on discounted subtotal
- Determine shipping based on discounted subtotal
- Add everything for final total

**Think about:**
- How will you iterate through the cart?
- There's a helper function above for quantity discounts - should you use it?
- How will you handle the membership discount, tax, and shipping calculations?
- What's the clearest way to structure this calculation?

**Example input:**
```python
cart = [
    {'item': 'Laptop', 'price': 800, 'quantity': 1, 'category': 'Electronics'},
    {'item': 'Mouse', 'price': 25, 'quantity': 2, 'category': 'Electronics'},
    {'item': 'Notebook', 'price': 5, 'quantity': 5, 'category': 'Supplies'},
]
is_member = True
```

````python
def calculate_shopping_cart_total(cart, is_member):
    """
    Calculate the final total for a shopping cart.
    
    Args:
        cart (list): List of item dicts with 'item', 'price', 'quantity', 'category'
        is_member (bool): Whether customer has membership
    
    Returns:
        float: Final total rounded to 2 decimal places
    """
    # Your code here
    total = 0
    for cart_item in cart:
        total += calculate_item_total(cart_item) 
    
    if is_member:
        total *= 0.85

    total *= 1.08

    # Shipping
    if total <= 100 and len(cart) > 0:
        total += 10
    
    return round(total, 2)
    
````

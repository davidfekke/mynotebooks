```python
def get_valid_input():
    while True:
        user_input = input("ENter a positive number: ")

        if user_input.isdigit() and int(user_input) > 0:
            print(f"You entered {user_input}")            
            break
            # return
        else:
            print("Invalid input")

get_valid_input()
print(f"You entered a valid number")
```

```python
def process_positive_numbers(numbers):
    index = 0
    results = []

    while index < len(numbers):
        num = numbers[index]
        index += 1

        if num <= 0:
            continue

        results.append(num)
    
    return results

my_list = [1, -3, 6, 8, -10, 16]
my_list2 = process_positive_numbers(my_list)
my_list2
```

```python
# List comprehension

numbers = [1, 2, 3]

squares = [num ** 2 for num in numbers]
print(squares)
```

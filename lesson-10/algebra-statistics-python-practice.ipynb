### Problem 2.3: Range Calculator with For Loop

Range shows us the spread of our data: how far apart the minimum and maximum values are.

**Practice:** Use a for loop to find the max and min values (don't use built-in max/min functions).

```python
def calculate_range(numbers):
    """
    Calculate the range (max - min) of a list.
    
    Args:
        numbers (list): List of numbers
    
    Returns:
        float: The range (difference between max and min)
    
    Challenge: Use a for loop to find max and min!
    """
    # Your code here
    my_min = numbers[0]
    my_max = numbers[0]
    for n in numbers:
        if n > my_max:
            my_max = n
        if n < my_min:
            my_min = n
    return my_max - my_min

# Test cases - DO NOT MODIFY
assert calculate_range([1, 5, 3, 9, 2]) == 8, "Range is 9 - 1 = 8"
assert calculate_range([10, 10, 10]) == 0, "All same values, range is 0"
assert calculate_range([-5, 5]) == 10, "Range is 5 - (-5) = 10"

print("✅ All tests passed!")
```

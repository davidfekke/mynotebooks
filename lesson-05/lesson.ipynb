```python
def add_numbers(a, b):
    return int(a + b)

def subtrack_numbers(a, b):
    return a - b


print(f"The numbers 4 + 6 = {add_numbers(4.1, 5.9)}")
print(f"The numbers 6 - 4 = {subtrack_numbers(b = 4, a = 6)}")
```

```python
total = add_numbers(3, 5) * 2
total
```

```python
def add_number_callback(a, b, cb):
    result = a + b
    cb(result)

def print_result(result) -> None:
    print(f"Your result is {result}")

add_number_callback(2, 5, print_result)

add_number_callback(6, 11, lambda res: print(res * 2))
```

## NEW: Line Charts for Time-Series Data

The Superstore dataset has **Order Date** - perfect for seeing trends over time!

Line charts are best when:
- You have **time-based data** (dates, months, years)
- You want to see **trends** (going up? down? seasonal patterns?)
- The x-axis has a **natural order** (unlike categories)

---

### Working with Dates in Pandas

First, we need to convert the 'Order Date' column to a proper date format:

```python
# Convert 'Order Date' to datetime format
df_superstore['Order Date'] = pd.to_datetime(df_superstore['Order Date'])

# Check the date range
print(f"Orders from: {df_superstore['Order Date'].min()}")
print(f"Orders to: {df_superstore['Order Date'].max()}")
```

### Example: Monthly Sales Trend

To see sales over time, we need to:
1. Group orders by month
2. Sum the sales for each month
3. Plot as a line chart

```python
# Create a 'Month' column (year-month format)
df_superstore['Month'] = df_superstore['Order Date'].dt.to_period('M')

# Group by month and sum sales
monthly_sales = df_superstore.groupby('Month')['Sales'].sum()

# Look at the first few months
print(monthly_sales.head(10))
```

```python
# Create a line chart of monthly sales
plt.figure(figsize=(14, 5))

# Convert period to string for plotting
x_values = monthly_sales.index.astype(str)
y_values = monthly_sales.values

plt.plot(x_values, y_values, color='steelblue', linewidth=2, marker='o', markersize=4)

plt.title('Monthly Sales Trend', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)

# Rotate x-axis labels so they don't overlap
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

### What Do You Notice?

Look at the pattern:
- Are there **peaks** at certain times of year? (Holiday shopping?)
- Is there an overall **upward or downward trend**?
- Any **unusual dips**?

### Challenge 6.3
**Compare category performance over time**

Create a line chart with **3 lines** - one for each category's monthly sales.

Steps:
1. Filter for each category (Furniture, Office Supplies, Technology) using `df['Product Category']`
2. For each filtered DataFrame, group by Month and sum Sales
3. Plot all 3 lines on the same chart
4. Add a legend to identify each line: `plt.legend(['Furniture', 'Office Supplies', 'Technology'])`

Which category consistently performs best?

```python
# Your code here


def add_plot(category, line_color):
    store_category = df_superstore[df_superstore['Product Category'] == category]
    mountly_sales_sum = store_category.groupby('Month')['Sales'].sum()
    x_values = mountly_sales_sum.index.astype(str)
    y_values = mountly_sales_sum.values

    plt.plot(x_values, y_values, color=line_color, linewidth=2, marker='o', markersize=4)

plt.figure(figsize=(14, 5))

# Convert period to string for plotting # .astype(str)
add_plot('Office Supplies', "steelblue")
add_plot('Furniture', "green")
add_plot('Technology', "red")

plt.title('Monthly Profit Trend', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=9)
plt.ylabel('Total Profit ($)', fontsize=12)

# Rotate x-axis labels so they don't overlap
plt.xticks(rotation=60, ha='right')

plt.legend(['Office Supplies', 'Furniture', 'Technology'])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (x vs y)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Compute slope (a) and intercept (b) using formulas
print(f"Covariance {np.cov(x, y, bias=True)}")
print(f"Covariance [0, 1] {np.cov(x, y, bias=True)[0, 1]}")
print(f"Variance {np.var(x)}")
a = np.cov(x, y, bias=True)[0, 1] / np.var(x)
b = np.mean(y) - a * np.mean(x)
print(f"Mean of y {np.mean(y)}")
print(f"Mean of x: {np.mean(x)}")
print("Slope:", a)
print("Intercept:", b)

# Predicted line
print(f"a: {a} * x: {x} + b: {b}")
print(f"a * x: {a * x} + b: {b}")
y_pred = a * x + b
print(f"y_pred: {y_pred}")

# Plot
plt.scatter(x, y, color='red')
plt.plot(x, y_pred)
plt.show()
```

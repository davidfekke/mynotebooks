```python
import pandas as pd
import numpy as np

df = pd.read_csv("./employee_attrition.csv")
df.shape
```

```python
# Feature 1: Promotion Ratio fraction of time gone without promotion
df['PromotionRatio'] = df['YearsSinceLastPromotion'] / (df['YearsAtCompany'] + 1 )
df
```

```python
# Feature 2: Are they getting paid fairly? IncomePerYearExp
df['IncomePerYearExp'] = (df['MonthlyIncome']) / (df['TotalWorkingYears'] + 1) 
df['IncomePerYearExp']
```

```python
# Feature transformation
# Take square root of every single value
df['SqrtTraining'] = df['TrainingTimesLastYear'] ** 0.5
df['SqrtTraining']
df.head()
```

```python
# One hot encoding
df['Department'].value_counts()
```

```python
department_encoded = pd.get_dummies(df['Department'], prefix='Dept')
department_encoded.head()
```

```python
df = pd.concat([df, department_encoded], axis=1)
df
```

```python
df['OverTime_Yes'] = (df['OverTime'] == 'Yes').astype(int)
df
```

```python
# Binning
age_bins = [18, 25, 35, 45, 55, 100]
age_labels = ['18-25', '26-35', '36-45', '46-55', '56+']

df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
df
```

```python
df['AgeGroup'].value_counts()
```

```python
# Normalizing
# df['MonthlyIncome'].min()
# df['MonthlyIncome'].max()
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()

df['Income_Normalized'] = minmax_scaler.fit_transform(df[['MonthlyIncome']])
df
```
